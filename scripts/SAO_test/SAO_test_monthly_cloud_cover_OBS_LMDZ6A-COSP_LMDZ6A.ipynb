```python
# OBS
path = '/home/jcharrel/Documents/M2_Internship_LMD/data/CALIPSO/'
ncfile_obs=path+"CALIPSO_2DCloudCover_seasonal_2008-2022_CFMIP2_sat_3.1.4_r144x90.nc"
nc_obs = netCDF4.Dataset(ncfile_obs)
```

```python
# LMDZ6A FOR T500, SLP, Z500, PRW, PRLW, PRSW
path = '/home/jcharrel/Documents/M2_Internship_LMD/data/LMDZ6A/'
ncfile_lmdz6a=path+'LmdzOr6Actrl_SE_1980_1999_1M_histmth_slp_t500_z500_prw_prlw_prsw.nc'
nc_lmdz6a = netCDF4.Dataset(ncfile_lmdz6a)
```

```python
# LMDZ6A FOR CLHCALIPSO, CLMCALIPSO, CLLCALIPSO, CLTCALIPSO
path = '/home/jcharrel/Documents/M2_Internship_LMD/data/LMDZ6A/'
ncfile_lmdz6a_2=path+'LmdzOr6Actrl_calipso_r144x90.nc'
nc_lmdz6a_2 = netCDF4.Dataset(ncfile_lmdz6a_2)
```

```python
# LMDZ6A FOR CLDL, CLDM, CLDH, CLDT
path = '/home/jcharrel/Documents/M2_Internship_LMD/data/LMDZ6A/'
ncfile_lmdz6a_3=path+'LmdzOr6Actrl_SE_1980_1999_1M_histmth_cld.nc'
nc_lmdz6a_3 = netCDF4.Dataset(ncfile_lmdz6a_3)
```

```python
# GRID
lon = nc_lmdz6a.variables['lon'][:]
lat = nc_lmdz6a.variables['lat'][:]

lon_2 = nc_lmdz6a_2.variables['lon'][:]
lat_2 = nc_lmdz6a_2.variables['lat'][:]

lat_50S = lat[(lat >= -51) & (lat < -49.5)]
print(lat_50S)
lat_65S = lat[(lat >= -61) & (lat < -59.6)]
print(lat_65S)

lat_id_50S = lat == lat_50S
lat_id_65S = lat == lat_65S

lat_50S_2 = lat_2[(lat_2 >= -51) & (lat_2 < -49.5)]
print(lat_50S_2)
lat_65S_2 = lat_2[(lat_2 >= -61) & (lat_2 < -59.6)]
print(lat_65S_2)

lat_id_50S_2 = lat_2 == lat_50S_2
lat_id_65S_2 = lat_2 == lat_65S_2
```

```python
###############################################
# CLL, CLM, CLH, CLT
###############################################
clllmdz6a_50S = nc_lmdz6a_2.variables['cllcalipso'][:,lat_id_50S_2,:]
clllmdz6a_50S_mean = np.mean(clllmdz6a_50S, axis=2)
clllmdz6a_65S = nc_lmdz6a_2.variables['cllcalipso'][:,lat_id_65S_2,:]
clllmdz6a_65S_mean = np.mean(clllmdz6a_65S, axis=2)

clmlmdz6a_50S = nc_lmdz6a_2.variables['clmcalipso'][:,lat_id_50S_2,:]
clmlmdz6a_50S_mean = np.mean(clmlmdz6a_50S, axis=2)
clmlmdz6a_65S = nc_lmdz6a_2.variables['clmcalipso'][:,lat_id_65S_2,:]
clmlmdz6a_65S_mean = np.mean(clmlmdz6a_65S, axis=2)

clhlmdz6a_50S = nc_lmdz6a_2.variables['clhcalipso'][:,lat_id_50S_2,:]
clhlmdz6a_50S_mean = np.mean(clhlmdz6a_50S, axis=2)
clhlmdz6a_65S = nc_lmdz6a_2.variables['clhcalipso'][:,lat_id_65S_2,:]
clhlmdz6a_65S_mean = np.mean(clhlmdz6a_65S, axis=2)

cltlmdz6a_50S = nc_lmdz6a_2.variables['cltcalipso'][:,lat_id_50S_2,:]
cltlmdz6a_50S_mean = np.mean(cltlmdz6a_50S, axis=2)
cltlmdz6a_65S = nc_lmdz6a_2.variables['cltcalipso'][:,lat_id_65S_2,:]
cltlmdz6a_65S_mean = np.mean(cltlmdz6a_65S, axis=2)

cllobs_50S = nc_obs.variables['cllcalipso'][:,lat_id_50S_2,:]
cllobs_50S_mean = np.mean(cllobs_50S[:,0,:], axis=1)
cllobs_65S = nc_obs.variables['cllcalipso'][:,lat_id_65S_2,:]
cllobs_65S_mean = np.mean(cllobs_65S[:,0,:], axis=1)

clmobs_50S = nc_obs.variables['clmcalipso'][:,lat_id_50S_2,:]
clmobs_50S_mean = np.mean(clmobs_50S[:,0,:], axis=1)
clmobs_65S = nc_obs.variables['clmcalipso'][:,lat_id_65S_2,:]
clmobs_65S_mean = np.mean(clmobs_65S[:,0,:], axis=1)

clhobs_50S = nc_obs.variables['clhcalipso'][:,lat_id_50S_2,:]
clhobs_50S_mean = np.mean(clhobs_50S[:,0,:], axis=1)
clhobs_65S = nc_obs.variables['clhcalipso'][:,lat_id_65S_2,:]
clhobs_65S_mean = np.mean(clhobs_65S[:,0,:], axis=1)

cltobs_50S = nc_obs.variables['cltcalipso'][:,lat_id_50S_2,:]
cltobs_50S_mean = np.mean(cltobs_50S[:,0,:], axis=1)
cltobs_65S = nc_obs.variables['cltcalipso'][:,lat_id_65S_2,:]
cltobs_65S_mean = np.mean(cltobs_65S[:,0,:], axis=1)
###############################################
# CLDL, CLDM, CLDH, CLDT
###############################################
cldllmdz6a_50S = nc_lmdz6a_3.variables['cldl'][:,lat_id_50S,:]
cldllmdz6a_50S_mean = np.mean(cldllmdz6a_50S, axis=2)
cldllmdz6a_65S = nc_lmdz6a_3.variables['cldl'][:,lat_id_65S,:]
cldllmdz6a_65S_mean = np.mean(cldllmdz6a_65S, axis=2)

cldmlmdz6a_50S = nc_lmdz6a_3.variables['cldm'][:,lat_id_50S,:]
cldmlmdz6a_50S_mean = np.mean(cldmlmdz6a_50S, axis=2)
cldmlmdz6a_65S = nc_lmdz6a_3.variables['cldm'][:,lat_id_65S,:]
cldmlmdz6a_65S_mean = np.mean(cldmlmdz6a_65S, axis=2)

cldhlmdz6a_50S = nc_lmdz6a_3.variables['cldh'][:,lat_id_50S,:]
cldhlmdz6a_50S_mean = np.mean(cldhlmdz6a_50S, axis=2)
cldhlmdz6a_65S = nc_lmdz6a_3.variables['cldh'][:,lat_id_65S,:]
cldhlmdz6a_65S_mean = np.mean(cldhlmdz6a_65S, axis=2)

cldtlmdz6a_50S = nc_lmdz6a_3.variables['cldt'][:,lat_id_50S,:]
cldtlmdz6a_50S_mean = np.mean(cldtlmdz6a_50S, axis=2)
cldtlmdz6a_65S = nc_lmdz6a_3.variables['cldt'][:,lat_id_65S,:]
cldtlmdz6a_65S_mean = np.mean(cldtlmdz6a_65S, axis=2)
```

```python
fig, axes = plt.subplots(figsize=(16,9), nrows=2, ncols=2)

months=np.arange(0,12)
months = ['Jan', 'Fev', 'Mar', 'Avr', 'Mai', 'Jun', 'Jui', 'Aou', 'Sep', 'Oct', 'Nov', 'Dec']

axes[0,0].plot(months,cllobs_50S_mean, label='50S OBS', c='r')
axes[0,0].plot(months,clllmdz6a_50S_mean, label='50S LMDZ6A-COSP', c='r', linestyle='--')
axes[0,0].plot(months,cldllmdz6a_50S_mean, label='50S LMDZ6A', c='r', linestyle=':')
axes[0,0].plot(months,cllobs_65S_mean, label='65S OBS', c='b')
axes[0,0].plot(months,clllmdz6a_65S_mean, label='65S LDMZ6A-COSP', c='b', linestyle='--')
axes[0,0].plot(months,cldllmdz6a_65S_mean, label='65S LMDZ6A', c='b', linestyle=':')
axes[0,0].set_xticks(months,months)
axes[0,0].set_xlim('Jan','Dec')
axes[0,0].set_title('Monthly low cloud cover')
axes[0,0].set_ylabel('Cloud cover fraction')
#ax_diff = axes[0,0].twinx()
#ax_diff.plot(months, cll_50S_mean - cll_65S_mean, label='Difference', c='k', linestyle='--')
#ax_diff.set_ylabel('Difference')

axes[0,1].plot(months,clmobs_50S_mean, c='r')
axes[0,1].plot(months,clmlmdz6a_50S_mean, c='r', linestyle='--')
axes[0,1].plot(months,cldmlmdz6a_50S_mean, c='r', linestyle=':')
axes[0,1].plot(months,clmobs_65S_mean, c='b')
axes[0,1].plot(months,clmlmdz6a_65S_mean, c='b', linestyle='--')
axes[0,1].plot(months,cldmlmdz6a_65S_mean, c='b', linestyle=':')
axes[0,1].set_xticks(months,months)
axes[0,1].set_xlim('Jan','Dec')
axes[0,1].set_title('Monthly mid cloud cover')
axes[0,1].set_ylabel('Cloud cover fraction')
# ax_diff = axes[0,1].twinx()
# ax_diff.plot(months, clm_50S_mean - clm_65S_mean, label='Difference', c='k', linestyle='--')
# ax_diff.set_ylabel('Difference')

axes[1,0].plot(months,clhobs_50S_mean, c='r')
axes[1,0].plot(months,clhlmdz6a_50S_mean, c='r', linestyle='--')
axes[1,0].plot(months,cldhlmdz6a_50S_mean, c='r', linestyle=':')
axes[1,0].plot(months,clhobs_65S_mean, c='b')
axes[1,0].plot(months,clhlmdz6a_65S_mean, c='b', linestyle='--')
axes[1,0].plot(months,cldhlmdz6a_65S_mean, c='b', linestyle=':')
axes[1,0].set_xticks(months,months)
axes[1,0].set_xlim('Jan','Dec')
axes[1,0].set_title('Monthly high cloud cover')
axes[1,0].set_ylabel('Cloud cover fraction')
# ax_diff = axes[1,0].twinx()
# ax_diff.plot(months, clh_50S_mean - clh_65S_mean, label='Difference', c='k', linestyle='--')
# ax_diff.set_ylabel('Difference')

axes[1,1].plot(months,cltobs_50S_mean, c='r')
axes[1,1].plot(months,cltlmdz6a_50S_mean, c='r', linestyle='--')
axes[1,1].plot(months,cldtlmdz6a_50S_mean, c='r', linestyle=':')
axes[1,1].plot(months,cltobs_65S_mean, c='b')
axes[1,1].plot(months,cltlmdz6a_65S_mean, c='b', linestyle='--')
axes[1,1].plot(months,cldtlmdz6a_65S_mean, c='b', linestyle=':')
axes[1,1].set_xticks(months,months)
axes[1,1].set_xlim('Jan','Dec')
axes[1,1].set_title('Monthly total cloud cover')
axes[1,1].set_ylabel('Cloud cover fraction')
# ax_diff = axes[1,1].twinx()
# ax_diff.plot(months, clt_50S_mean - clt_65S_mean, label='Difference', c='k', linestyle='--')
# ax_diff.set_ylabel('Difference')

fig.tight_layout()
fig.legend(loc='center left', bbox_to_anchor=(1, 0.5))
path_fig = '/home/jcharrel/Documents/M2_Internship_LMD/figures/SAO_test/'
fig.savefig(path_fig+'SAO_monthly_cloud_cover_OBS_LMDZ6A-COSP_LMDZ6A.png',dpi=300,bbox_inches='tight')
```
